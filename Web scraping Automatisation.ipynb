### 2.8 Créez une fonction `scraping_bdm` qui collecte l'ensemble des données des articles présent sur la page d'acceuil et qui renvoi un fichier au format `json`.

```python
import requests
from bs4 import BeautifulSoup

# requete vers le site blog du modérateur
response2 = requests.get("https://www.blogdumoderateur.com/")
# interpretation du site par beautifulsoup
soup = BeautifulSoup(response2.text)
# recuperation des articles dans la variable articles
articles = soup.find_all("article")


# boucle sur les articles

for article in articles:
    try:
        # titre de l'article
        print(article.find("h3").text)
        # lien de l'image
        print(article.find("img").attrs["data-lazy-src"])
        # date de l'article
        print(article.find_all('span')[1].text)
        # catégorie de l'article
        print(article.find("span").text)
        # lien de l'article
        print(article.find("a").attrs['href'])
    except:
        pass
```

```python
# Création d'un fichier json
import json

data_json = {
    1 :{
    'title': "Actu ChatGPT",
    'img_url': "httts://blogdumoderateur.com/"
    }
}

# Ecriture d'un fichier JSON (sauvegarde)
with open('./data.json', 'w') as f:
    json.dump(data_json, f)

# Lecture d'un fichier JSON (lecture)
with open('./data.json', 'r') as f:
    data_load = json.load(f)

data_load
```
